```python
import numpy as np
import PIL.Image as Image
from io import BytesIO

from minitorch.tensor.tensor import Tensor
from minitorch.transform.transform import (
    RandomCrop, RandomFlip,
    Compose, RandomState,
    Normalization)
from minitorch.transform.helpers import ToTensor,to_chw
```

```python
image_path = 'C:\\Users\\User\\Pictures\\bunny-1845263_1920.jpg'
img = Image.open(image_path)
image_array = np.array(img)
image = to_chw(image_array)
```

```python
state = RandomState(seed=42)
flipped_image = RandomFlip(prob=0.5, state=state, axis=-1)(image)
cropped_image = RandomCrop(size=(10,10), padding=2, state=state)(flipped_image)
normalized_image = Normalization(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])(cropped_image)
image_tensor = ToTensor()(normalized_image)
```

```python
state = RandomState(seed=42)
transform = Compose(
    [RandomCrop(size=(100,100), padding=0, state=state),
    RandomFlip(prob=0.5),
    Normalization(mean=[0.5, 0.5, 0.5],
                std=[0.5, 0.5, 0.5]),
    ToTensor()]
)

transformed_image = transform(image)
```
